```python
X = df.drop(columns=['Exited'])
y = df['Exited']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

### Modeling

```python
def untuned(X_train, X_test, y_train, y_test):
    lr = LogisticRegression()
    knn = KNeighborsClassifier()
    dt = DecisionTreeClassifier()
    rf = RandomForestClassifier()
    ada = AdaBoostClassifier()
    lgb = LGBMClassifier()
    xgb = XGBClassifier()
    cat = CatBoostClassifier()
    
    names = ['Linear Regression', 'KNN', 'DT', 'RF', 'AdaBoost', 'Light GBM', 'XGBoost', 'CatBoost']
    algos = [lr, knn, dt, rf, ada, lgb, xgb, cat]
    acc = [None] * 8
    precision = [None] * 8
    recall = [None] * 8
    f1 = [None] * 8
    
    for indx, model in enumerate(algos):
        try:
            if(indx < 7):
                model.fit(X_train, y_train)
            else:
                model.fit(X_train, y_train, silent=True)
                
            y_pred = model.predict(X_test)
            acc[indx] = model.score(X_test, y_test)            
            precision[indx] = precision_score(y_test, y_pred)
            recall[indx] = recall_score(y_test, y_pred)
            f1[indx] = f1_score(y_test, y_pred)
            
        except Exception as e:
            acc[indx] = "-"
            precision[indx] = "-"
            recall[indx] = "-"
            f1[indx] = "-"
            print(f"{names[indx]}: {str(e)}")
        
    return pd.DataFrame([acc, precision, recall, f1],
                        index=['Accuracy', 'Precision', 'Recall', 'F1'],columns=names).T
```

```python
nothing = untuned(X_train, X_test, y_train, y_test)
nothing
```

Observation: None of the models work well with non-encoded categorical

Let's perform one-hot encoding 

```python
columns = ['Gender', 'Geography']
for col in columns:
    one_hot = pd.get_dummies(df[col], drop_first=True)
    df = df.drop(columns=[col])
    df = df.join(one_hot)
    
df.info()

X = df.drop(columns=['Exited'])
y = df['Exited']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
encoding = untuned(X_train, X_test, y_train, y_test)
encoding
```

```python
ss = StandardScaler()
X = ss.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
scaling = untuned(X_train, X_test, y_train, y_test)
scaling
```

We can notice that Catboost, XGBoost, Light GBM, Random Forest, and Decision tree are robust to scale. As they do not use distance. Meanwhile, KNN and Linear Regression have improved.

```python
sm = SMOTE(random_state=42)
X, y = sm.fit_resample(X, y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
imbalance = untuned(X_train, X_test, y_train, y_test)
imbalance
```

We can see that after balancing the dataset, the precision, recall, F1 scores increased drastically in all algorithms. They are all sensitive to imbalance.
